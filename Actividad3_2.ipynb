```python
#Carga desde un archivo .xlsx sin indice
df1 = pd.read_excel('Datos de Facturaci√≥n.xlsx')
df1
```

```python
df1.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
#Hacer copia
df11 = df1.copy()
```

```python
# Dropear columna (Fecha_Cancela)
df11 = df1.drop(['FECHA_CANCELA'], axis = 1)
df11

# Se elimina la columna porque la gran mayor√≠a de los datos son nulos desvirtuando analisis
```

```python
#Sustituir valores nulos con "forward fill"
df11["FECHAELAB"] =df1["FECHAELAB"].fillna(method="ffill")
df11["FECHA_DOC"] =df1["FECHA_DOC"].fillna(method="ffill")
df11["FECHA_ENT"] =df1["FECHA_ENT"].fillna(method="ffill")
df11["CVE_VEND"] =df1["CVE_VEND"].fillna(method="ffill")
df11["STATUS"] =df1["STATUS"].fillna(method="ffill")

# Se sustituye por los valores siguientes de las filas ya que se sigue un patr√≥n consecuente en las columnas
```

```python
# Sustituir por registro √∫nico de identificaci√≥n

df11["CVE_DOC"] =df1["CVE_DOC"].fillna("F00000")

# Se sustituyen los nulos de registro siendo intercambiados por una clave de identificaci√≥n √∫nica
```

```python
#Identificar valores nulos por columna
valores_nulos=df11.isnull().sum()
valores_nulos
```

```python
# Sustituir por mediana

df11["DES_TOT"]= df11["DES_TOT"].fillna(round(df1["DES_TOT"].median(),1))

df11["DES_FIN"]= df11["DES_FIN"].fillna(round(df1["DES_FIN"].median(),1))
df11

# Se sustituye por la mediana ya que la gran mayor√≠a de valores comparten n√∫mero siendo en este caso el "0"
```

```python
# Sustituir por promedio

df11["CAN_TOT"]= df11["CAN_TOT"].fillna(round(df1["CAN_TOT"].mean(),1))
df11

# La cantidad en pago se rellena con el promedio de las facturas para no desvirtuar pr√≥ximos an√°lisis
```

```python
#Identificar valores nulos por columna
valores_nulos=df11.isnull().sum()
valores_nulos
```

```python
df2 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
```

```python
df2.info()
```

```python
#Identificar valores nulos por columna
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Hacer copia
df22 = df2.copy()
```

```python
# Sustituir por registro √∫nico de identificaci√≥n

df22["CVE_DOC"] =df2["CVE_DOC"].fillna("F00000")
df22["NOMBRE_VENDEDOR"] =df2["NOMBRE_VENDEDOR"].fillna("Vendedor No Identificado")
df22["NOMBRE_CLIENTE"] =df2["NOMBRE_CLIENTE"].fillna("Cliente No Identificado")
df22["CVE_ART"] =df2["CVE_ART"].fillna("Cliente No Identificado")
df22["DESCR"] =df2["DESCR"].fillna("Producto No Identificado")

# Se sustituyen los nulos de registro siendo intercambiados por una clave de identificaci√≥n √∫nica
```

```python
#Identificar valores nulos por columna
valores_nulos=df22.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos con "forward fill"
df22["FECHA_DOC"] =df2["FECHA_DOC"].fillna(method="ffill")

# Se sustituye por los valores siguientes de las filas ya que se sigue un patr√≥n consecuente en las columnas
```

```python
#Identificar valores nulos por columna
valores_nulos=df22.isnull().sum()
valores_nulos
```

```python
# Sustituir por mediana

df22["CANT"]= df22["CANT"].fillna(round(df2["CANT"].median(),1))
df22["PRECIO_UNITARIO"]= df22["PRECIO_UNITARIO"].fillna(round(df2["PRECIO_UNITARIO"].median(),1))
df22["COSTO_UNITARIO"]= df22["COSTO_UNITARIO"].fillna(round(df2["COSTO_UNITARIO"].median(),1))
df22["COSTO_UNITARIO_CALCULADO"]= df22["COSTO_UNITARIO_CALCULADO"].fillna(round(df2["COSTO_UNITARIO_CALCULADO"].median(),1))
df22["SUBTOTAL_PARTIDA"]= df22["SUBTOTAL_PARTIDA"].fillna(round(df2["COSTO_UNITARIO"].median(),1))

df22

# Se sustituye por la mediana ya que la gran mayor√≠a de valores comparten n√∫mero siendo en este caso el "0"
```

```python
#Identificar valores nulos por columna
valores_nulos=df22.isnull().sum()
valores_nulos
```
